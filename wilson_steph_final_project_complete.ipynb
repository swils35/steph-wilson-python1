```python
import csv
import matplotlib.pyplot as plt

schools_listed = []
average_lines = []
lunch = []

# call all other functions and print the updated data
def main():
            
        percent = percent_in_program()   
        average_prof = averages()
        school = school_names()
        
        print("Below is a table which shows the proficiency average of each school and its percentage of students")
        print("in a lunch program. You will also find a scatterplot depicting the correlation of the two variables.")
        print()
        
        
        print("% in lunch    % Proficiency   School")
        print(" program \t  Average")
        print("------------------------------------------")
    
        table()
        
        chart()
     
        
        
# calculate the average proficiency for each school by adding up the math and
# reading proficiencies then dividing by two
def averages():    
    
    with open("final_data_for_averages.csv") as averages_data:
        reader = csv.reader(averages_data, delimiter = ",")   
        
        for line in reader:
            
            line_sum = sum(float(d) for d in line)
            av_lines = (line_sum / 2 * 100)
            av_lines = (format(av_lines, ".2f") + "%")
            average_lines.append(av_lines)
            
            
            
        return average_lines



# returning the lunch program data from the final_project_data
# file in list form to be used in table function
def percent_in_program():
    
    with open ("final_project_data.csv", newline="") as lunch_programs:
        reader = csv.DictReader(lunch_programs) 
              
        for row in reader:
            lunch_programs = row["Free/ Reduced-Price Lunch Program"]
            lunch.append(lunch_programs)
            
            
        return lunch

        

# returning the school names data from the final_project_data
# file in list form to be used in table function        
def school_names():
    
    with open ("final_project_data.csv", newline="") as school_list:
        reader = csv.DictReader(school_list)
        
        for row in reader:
            schools = row["Schools"]
            schools_listed.append(schools)
            
        return schools_listed 



# creating a table with the data from the averages function,
# as well as the lists created in the percent_in_program and
# school_names functions.
def table():
        
    for x,y,z in zip(lunch, average_lines, schools_listed):
        print(" " + x + "\t\t", y + "\t    ", z)
        
    

# creating a chart to show the correlation between
# the two variables.
def chart():
    
    x = []
    y = []
    
    with open("final_project_chart_data.csv", "r", encoding = "utf-8") as chart_data:
        plots = csv.reader(chart_data, delimiter=",")
        
        for row in plots:
            x.append(float(row[0]))
            y.append(float(row[1]))
            
    plt.plot(x,y, label="", linestyle="", marker="o")
    plt.xlabel("Percentage of students \n in a lunch program")
    plt.ylabel("Average Math & Reading \n Proficiency %")
    plt.title("Correlation Between Income and \n Average Proficiency\n in Standardized Testing")
    plt.show()
    
    
    
main()
```
