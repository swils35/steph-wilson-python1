```python
import csv

# call all other functions and print the updated data
def main():
            
        percent = percent_in_program()   
        average_prof = averages()
        school = school_names()
        
        table(percent, average_prof, school)
        
       # print("% in lunch \t % Proficiency \t \t School")
       # print("program \t Average")
       # print("----------------------------------------")
     
        
        

# calculate the average proficiency for each school by adding up the math and
# reading proficiencies then dividing by two
def averages():    
    
    with open("final_data_for_averages.csv") as averages_data:
        reader = csv.reader(averages_data, delimiter = ",")   
        
        average_lines = [] # create an empty list to put averages into
        
        for line in reader:
            
            line_sum = sum(float(d) for d in line)
            av_lines = (line_sum / 2 * 100)
            av_lines = (format(av_lines, ".2f") + "%")
            average_lines.append(av_lines)
            
            
            
        return average_lines




# returning the lunch program data from the final_project_data
# file in list form to be used in table function
def percent_in_program():
    
    with open ("final_project_data.csv", newline="") as lunch_programs:
        reader = csv.DictReader(lunch_programs) 
        
        lunch = []
           
        for row in reader:
            lunch_programs = row["Free/ Reduced-Price Lunch Program"]
            lunch.append(lunch_programs)
            
            
        return lunch

        

# returning the school names data from the final_project_data
# file in list form to be used in table function        
def school_names():
    
    with open ("final_project_data.csv", newline="") as school_list:
        reader = csv.DictReader(school_list)
        
        schools_listed = []
        
        for row in reader:
            schools = row["Schools"]
            schools_listed.append(schools)
            
        return schools_listed 



# creating a new csv file with the data from the averages function,
# as well as the lists created in the percent_in_program and
# school_names functions to be used to make a prettier display table.
def table(percent, average, schools):
    
    with open ("final_project_new_data.csv", "w", newline="") as new_data_file:
        writer = csv.writer(new_data_file, dialect="excel")

        # gotta figure out how to write each list to its own COLUMN.
        writer.writerow(percent)
        writer.writerow(average)
        writer.writerow(schools)
    

    
# function that will create a chart using the average proficiency
# and lunch program data.
# def chart():
    
    
    
main()
```
