The program should:

1. Generate a random number between 1 and 100 (without showing it). 
2. Prompt the user to guess the generated number.
    1. If correct, congratulate them and tell them how many guesses it took, then ask if they wish to play again. If wrong, ask them to try again. 
    
main function: Introduces the game and sets up play_again loop to True in order to start play function.
<br/>get_guess function: Asks the user to guess the number and validates the input.
<br/>play function: generates the computer's random number then compares the user's guess against said random number and prints responses. If guess correct, congratulates and asks user if they wish to play again. If so, play_again loop in main begins again.

```python
import random

def main():       
    
    print("I'm thinking of a number between 1 and 100.")
    
    play_again = True 
    
    while play_again:
        play_again = play() # call play function
    
    
def get_guess():
    guess = int(input("Guess the number that I'm thinking of: ")) # prompt user
    
    while 1 > guess or guess > 100:  # validate guess
        print("Sorry! Apparently failure IS an option.")
        guess = int(input("Please enter a valid guess. "))
        
    return guess

def play():
    count = 0 # initiate count for number of input
    guess = 0 # create variable to hold guesses
    rand_number = random.randint(1, 100) # generate random number
    
    while rand_number != guess:
        count += 1
        guess = get_guess()
        if rand_number == guess:
            print("Congratulations! You guessed correctly in", count, "attempt(s).")            
     
        elif rand_number > guess:
            print("Too low! Try again.")
            
        else:
            print("Too high! Try again.")
        
    again = input("Would you like to play again?" + 
                " If yes, press 'Y'. If no, press anything else. ")
    
    if "y" in again or "Y" in again: # switch for play_again loop in main
        return True
    else:
        return False
        
main()
```
